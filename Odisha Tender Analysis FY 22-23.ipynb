```python
import pandas as pd
```

```python
# Load the dataset
tender_data = pd.read_csv('assignment_sheet_odisha_tender_data.csv')
```

```python
# Display the dataframe
tender_data
```

```python
tender_data.shape
```

```python
# Define relevant keywords for flood management related tenders
keywords = ["flood", "disaster management", "infrastructure", 
            "embankment", "culvert", "drainage", "pokhari", "pond", "river", "bridge"]
```

```python
# Convert Title and Work Description columns to lowercase for case-insensitive search
tender_data['Title'] = tender_data['Title'].str.lower()
tender_data['Work Description'] = tender_data['Work Description'].str.lower()
```

```python
# Apply keywords filter
flood_related_tenders = tender_data[
    tender_data['Title'].str.contains('|'.join(keywords)) |
    tender_data['Work Description'].str.contains('|'.join(keywords))
]
```

```python
# Reset the index to keep the original row index as a new column
flood_related_tenders = flood_related_tenders.reset_index()

# Rename the index column for clarity
flood_related_tenders.rename(columns={'index': 'Original Index'}, inplace=True)
```

```python
# Display flood related tenders dataframe
flood_related_tenders
```

```python
# Reload the original dataset to get Title and Work Description in original case
tender_data_original = pd.read_csv('assignment_sheet_odisha_tender_data.csv')
```

```python
tender_data_original
```

```python
# Restore original Title and Work Description into the filtered flood-related tenders
flood_related_tenders['Title'] = tender_data_original['Title'].iloc[flood_related_tenders.index]
flood_related_tenders['Work Description'] = tender_data_original['Work Description'].iloc[flood_related_tenders.index]
```

```python
flood_related_tenders
```

```python
# Export the flood-related tenders to a CSV file
flood_related_tenders.to_csv('flood_related_tenders.csv', index=False)
```

```python
# Count tenders by department
department_counts = flood_related_tenders['Organisation Chain'].value_counts()
```

```python
# Convert the count to a aataframe for easy visualization
department_counts_df = department_counts.reset_index()
department_counts_df.columns = ['Department', 'Tender Count']
```

```python
department_counts_df
```

```python
# Export department counts dataframe to a CSV
department_counts_df.to_csv('department_counts.csv', index=False)
```

```python
# Save Top 5 departments by tender count in a dataframe
top_5_departments = department_counts_df.head(5)
```

```python
# Display Top 5 departments dataframe
top_5_departments
```

```python
# Count project types by Tender Category
category_counts = flood_related_tenders['Tender Category'].value_counts()
```

```python
# Convert to dataframe for easy visualization
category_counts_df = category_counts.reset_index()
category_counts_df.columns = ['Category', 'Count']
```

```python
category_counts_df
```

```python
# Convert 'Tender Value in â‚¹' to numeric by removing commas and converting to float
flood_related_tenders['Tender Value in â‚¹'] = flood_related_tenders['Tender Value in â‚¹'].str.replace(',', '').astype(float)

# Standardize 'Location' to lowercase for consistent grouping
flood_related_tenders['Location'] = flood_related_tenders['Location'].str.lower()

# Group by 'Location' and calculate both the total Tender Value and Tender Count
region_funding = flood_related_tenders.groupby('Location').agg(
    Tender_Value=('Tender Value in â‚¹', 'sum'),
    Tender_Count=('Tender Value in â‚¹', 'size')  # Counts the number of entries per location
).reset_index()

# Sort by total funding value
region_funding_sorted = region_funding.sort_values(by='Tender_Value', ascending=False)

# Capitalize the 'Location' names
region_funding_sorted['Location'] = region_funding_sorted['Location'].str.title()

# Format 'Tender_Value' with commas for rupee format
region_funding_sorted['Tender_Value'] = region_funding_sorted['Tender_Value'].apply(lambda x: f"{x:,.2f}")

# Rename columns for clarity
region_funding_sorted.rename(columns={'Tender_Value': 'Tender Value in â‚¹', 'Tender_Count': 'Tender Count'}, inplace=True)

# Export funding by region dataframe to a CSV
region_funding_sorted.to_csv('flood_related_region_funding_sorted.csv', index=False)
```

```python
# Convert Awarded Value to numeric
flood_related_tenders['Awarded Value'] = flood_related_tenders['Awarded Value'].str.replace(',', '').astype(float)
```

```python
# Sort by Awarded Value to find large-scale projects
large_projects = flood_related_tenders.sort_values(by='Awarded Value', ascending=False)
```

```python
# Convert Awarded Value to rupee
large_projects['Awarded Value'] = large_projects['Awarded Value'].apply(lambda x: f"{x:,.2f}")
```

```python
# Export sorted large projects dataframe to CSV
large_projects.to_csv('large_projects.csv', index=False)
```

```python
# Identify departments that handle the largest budgets for flood related projects
department_spending = flood_related_tenders.groupby('Organisation Chain')['Awarded Value'].sum().reset_index()
department_spending_sorted = department_spending.sort_values(by='Awarded Value', ascending=False)
```

```python
# Export department spending dataframe to CSV
department_spending_sorted.to_csv('department_spending.csv', index=False)
```
